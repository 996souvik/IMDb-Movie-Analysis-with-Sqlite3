```python
#Import lib function
import sqlite3
```

```python
#Create a new database
connection = sqlite3.connect("IMDb_Dataset.db")
print("Connected with database succesfully")
```

```python
#Cheak the connection type
print(type(connection))
```

```python
sqlstr = '''
        CREATE TABLE imdb_data(
        Title TEXT PRIMARY KEY,
        IMDb_Rating REAL,
        Year INTEGER,
        Certificates TEXT,
        Genre TEXT,
        Director TEXT,
        Star_Cast TEXT,
        MetaScore REAL,
        Duration REAL);
        '''

#Cheak if the table is created or not
try:
    connection.execute(sqlstr)
    print("Crate table succesfully....!")
except:
    print("Error!!!....")
```

```python
# SQL insert statement with OR IGNORE to skip duplicates
sqlstr = "INSERT OR IGNORE INTO imdb_data (Title, IMDb_Rating, Year, Certificates, Genre, Director, Star_Cast, MetaScore, Duration) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?);"

# Read lines from CSV file and store them into table
with open('IMDb_Dataset.csv', encoding='utf-8') as data_file:
    csv_reader = csv.reader(data_file)
    header = next(csv_reader)  # Skip the header row
    for row in csv_reader:
        connection.execute(sqlstr, (row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8]))
        
connection.commit() #Final execution
print("All recrods inserted successfully!!!")      
```

```python
sqlstr = """SELECT Year, COUNT(*) AS Movie_Count
            FROM imdb_data
            GROUP BY Year 
            ORDER BY Year;"""
#Read the table
cur_table = connection.execute(sqlstr)

#print the data 
for row in cur_table:
    print(f"Year: {row[0]}, Movie Count: {row[1]}")
    
```

<h2 style="color:#ff0000;font-size:30px;font-family:initial;text-align:center;"><strong>3.<strong style="color:#005aac;font-size:30px;font-family:initial;"><b>Find the Top 10 Actors/Actresses by Number of Movies.</b><strong style="color:#ff0000;font-size:30px;font-family:initial;">:-</strong></strong></strong></h2>

```python
sqlstr = """SELECT Star_Cast, COUNT(*) AS Movie_Count 
            FROM imdb_data 
            GROUP BY Star_Cast 
            ORDER BY Movie_Count DESC 
            LIMIT 10"""

#Read the table
cur_table = connection.execute(sqlstr)

#print the data 
for row in cur_table:
    print(f"Actor/Actress Name: {row[0]}, No of Movies: {row[1]}")
```

<h2 style="color:#ff0000;font-size:30px;font-family:initial;text-align:center;"><strong>4.<strong style="color:#005aac;font-size:30px;font-family:initial;"><b>Compare Average IMDb Ratings of Movies Before and After 2000.</b><strong style="color:#ff0000;font-size:30px;font-family:initial;">:-</strong></strong></strong></h2>

```python
sqlstr = """SELECT 
            CASE 
            WHEN Year < 2000 THEN 'Before 2000'
            ELSE '2000 and After'
            END AS Period, 
            AVG(IMDb_Rating) AS Avg_Rating
            FROM imdb_data 
            GROUP BY Period;
            """

#Read the table
cur_table = connection.execute(sqlstr)

#print the data 
for row in cur_table:
    print(f"before and after 2000: {row[0]}, IMDB RATING: {row[1]}")
```

<h2 style="color:#ff0000;font-size:30px;font-family:initial;text-align:center;"><strong>4.<strong style="color:#005aac;font-size:30px;font-family:initial;"><b>Compare the Average IMDb Rating and MetaScore of Movies.</b><strong style="color:#ff0000;font-size:30px;font-family:initial;">:-</strong></strong></strong></h2>

```python
sqlstr = """SELECT Title, IMDb_Rating, MetaScore, 
            CASE 
            WHEN IMDb_Rating > MetaScore THEN 'Higher IMDb Rating'
            WHEN IMDb_Rating < MetaScore THEN 'Higher MetaScore'
            ELSE 'Equal'
            END AS Comparison 
            FROM imdb_data 
            ORDER BY Title;
            """

#Read the table
cur_table = connection.execute(sqlstr)

#print the data 
for row in cur_table:
    print(f"MetaScore: {row[0]}, IMDB RATING: {row[1]}")
```

<h2 style="color:#ff0000;font-size:30px;font-family:initial;text-align:center;"><strong>5.<strong style="color:#005aac;font-size:30px;font-family:initial;"><b>Insert N  Number of Movies.</b><strong style="color:#ff0000;font-size:30px;font-family:initial;">:-</strong></strong></strong></h2>

```python
# Get the number of new movies from the user
n = int(input("Enter the number of new movies: "))
cursor = connection.cursor()
sqlstr = "INSERT INTO imdb_data VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?);"

# Loop to get movie details from the user
for i in range(n):
    Title = input("Enter Title (TEXT): ")
    IMDb_Rating = float(input("Enter IMDb Rating (REAL): "))
    Year = int(input("Enter Year (INTEGER): "))
    Certificates = input("Enter Certificates (TEXT): ")
    Genre = input("Enter Genre (TEXT): ")
    Director = input("Enter Director name (TEXT): ")
    Star_Cast = input("Enter Star Cast (TEXT): ")
    MetaScore = float(input("Enter MetaScore (REAL): "))
    Duration = float(input("Enter Duration (REAL): "))

    # Execute the SQL statement with the provided values
    cursor.execute(sqlstr, (Title, IMDb_Rating, Year, Certificates, Genre, Director, Star_Cast, MetaScore, Duration))

# Commit the transaction
connection.commit()
print("All records successfully uploaded to the database.")

```
